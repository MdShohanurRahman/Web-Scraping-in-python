````python
import requests
from bs4 import BeautifulSoup
from requests_html import HTMLSession
import re
import pandas as pd
````

````python
session = HTMLSession()
````

````python
# empty list
hadith_no = []
hadith_narrators = []
hadith_descriptions = []
hadith_similarity = []
hadit_levels = []
````

Quistin and answer pattern in website

````python
hadith_no_pattern = 'div.had>h3.hidden-print'
hadith_narrator_pattern = '.narrated-by'
hadith_description_pattern = '.hadith-des'
hadith_level_pattern = '.label.validity'
````

````python
r = session.get('http://ihadis.com/books/bukhari/chapter/1')
no = r.html.find(hadith_no_pattern)
narrator = r.html.find(hadith_narrator_pattern)
desc = r.html.find(hadith_description_pattern)
level = r.html.find(hadith_level_pattern)
````

````python
len(no) == len(desc)==len(narrator) == len(level)
````

````python
print(desc[0].text)
````

````python
for j, k, l, m in zip(no,narrator,desc,level):
    hadith_no.append(j.text)
    hadith_narrators.append(k.text)
    desc_sim = l.text.split('\n\n')
    hadith_descriptions.append(desc_sim[0])
    hadith_similarity.append(desc_sim[-1])
    hadit_levels.append(m.text)
````

````python
# empty list
hadith_no = []
hadith_narrators = []
hadith_descriptions = []
hadith_similarity = []
hadit_levels = []

for i in range(1,61):
    r = session.get('http://ihadis.com/books/muatta-malik/chapter/{}'.format(i))
    no = r.html.find(hadith_no_pattern)
    narrator = r.html.find(hadith_narrator_pattern)
    desc = r.html.find(hadith_description_pattern)
    level = r.html.find(hadith_level_pattern)
    
    if len(no) == len(desc)==len(narrator) == len(level):
        for j, k, l, m in zip(no,narrator,desc,level):
            hadith_no.append(j.text)
            hadith_narrators.append(k.text)
            desc_sim = l.text.split('\n\n')
            hadith_descriptions.append(desc_sim[0])
            hadith_similarity.append(desc_sim[-1])
            hadit_levels.append(m.text)
        
    
````

````python
len(hadith_no), len(hadith_narrators), len(desc_sim), len(hadith_descriptions), len(hadith_similarity), len(hadit_levels)
````

## Csv file datasets

````python
# create dataframe
data = {
    "Serial No" : hadith_no,
    "Narrator" : hadith_narrators,
    "Description" : hadith_descriptions,
    "Similarity" : hadith_similarity,
    "Level" : hadit_levels,
}
````

````python
df = pd.DataFrame(data,columns=['Serial No','Narrator','Description','Similarity','Level'])
df.to_csv('dara/my-scrap.csv',index=None, encoding='utf-8')
````
